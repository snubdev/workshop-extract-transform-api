```python
import pandas as pd
import datetime
```

```python
file_name = 'data.csv'
```

```python
# Carrega os dados do CSV para um DataFrame
df = pd.read_csv(file_name)
```

```python
# Exibe as primeiras linhas
df
```

```python
# Converte a coluna 'Data_Registro' para datetime, detectando formatos mistos e assumindo que o primeiro número é o dia
# É uma função do pandas que converte uma coluna ou série em objetos datetime do Python
# mixed indica que a coluna pode ter vários formatos de data diferentes
# dayfirst diz que o primeiro número é o dia (dd/mm/yyyy) em vez do mês

format_date = pd.to_datetime(df['Data_Registro'], format='mixed', dayfirst=True)
```

```python
# Insere substitui os formatos de data para os novos formatos
df['Data_Registro'] = format_date
```

```python
# Exibe as primeiras linhas
df.head()
```

```python
# Converte o formato data no formato BR
formart_data_br = format_date.dt.strftime('%d/%m/%Y')
```

```python
# Exibe os dados formatados
formart_data_br
```

```python
# Remove a Coluna ID
df = df.drop(columns=['ID'])
```

```python
# Exibe as primeiras linhas
df.head()
```

```python
# Marcar todos os dados duplicatos (incluindo a primeira)
duplicated_mask = df.duplicated(subset=["Nome_Produto", "Categoria", "Preco"], keep=False)
```

```python
# Filtrar o DataFrame
df_duplicated = df[duplicated_mask]
```

```python
# Exibe um novo DataFrama de Dados Duplicado
df_duplicated
```

```python
cont = 0
for i in df["Categoria"]:
    if "Eletrônicos" == i:
        cont += 1
print(cont)
```

```python
# Usando value_counts()
# Conta quantas vezes o valor aparece
cont_v2 = df["Categoria"].value_counts().get("Eletrônicos", 0)
```

```python
# Cria um novo arquivo csv sem dados duplicados
df.to_csv('formatted_data.csv', index=False)
```
