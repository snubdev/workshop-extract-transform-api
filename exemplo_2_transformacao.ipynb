```python
import requests
import pandas as pd
import datetime
```

```python
url = 'https://viacep.com.br/ws/BA/Feira de Santana/SIM/json/'
```

### Fazendo a requisição GET para a API

```python
response = requests.get(url)
```

### Convertendo o JSON retornado em uma lista de dicionários

```python
lista_ceps = response.json()
```

```python
lista_ceps
```

### Transformando em um DataFrame do Pandas

```python
df = pd.DataFrame(lista_ceps)
```

### Exibindo as primeiras linhas do DataFrame

```python
df.head()
```

### Exibindo as Últimas linhas do DataFrame

```python
df.tail()
```

### Exibindo a estrutura do DataFrame

```python
df.info()
```

### Exibindo as colunas disponíveis no DataFrame

```python
print("Colunas do DataFrame:", df.columns)
```

### Adicionando uma coluna com a data e hora da coleta

```python
df['data_coleta'] = datetime.datetime.now()
```

```python
df.head()
```

### Removendo colunas que não serão utilizadas para análise

```python
df = df.drop(columns=['complemento', 'unidade','estado','ibge', 'gia', 'ddd', 'siafi'])
```

### Exibindo as primeiras linhas do DataFrame atualizado

```python
print("DataFrame após remoção de colunas:")
df.head()
```

### Criando uma nova coluna 'regiao' com o nome da cidade em maiúsculo

```python
df['regiao'] = df['localidade'].str.upper()
```

### Exibindo o DataFrame atualizado com a nova coluna

```python
print("DataFrame com coluna 'regiao' em maiúsculo:")
df.head()
```
